## Data Cleaning

See https://pandas.pydata.org/pandas-docs/stable/getting_started/10min.html for an introduction to Pandas.

```python
# we need to import the libraries before we may use them
# if an ImportError is thrown, you need to install the libraries via pip3
import numpy as np
import pandas as pd
# both are now accessible using np respectively pd before the function name
```

We need to import and clean the data before we may explore it.
We use the dataset from https://www.kaggle.com/rhuebner/human-resources-data-set.

```python
df = pd.read_csv('HRDataset_v14.csv', nrows=310, index_col='EmpID')
# we use index_col=EmpID to use the date as index and not the line number
# we use nrows=310 to avoid the empty lines at the bottom of the file

# Pandas uses DataFrame which is like a spreadsheet table
df
```

```python
# we can get basic metrics of numerical columns to understand the data
df.describe()
```

```python
# check missing values
df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
# TermReason, DateofTermination, EmpStatusID and EmploymentStatus hold information regarding termination
# we remove them because for a prediction they are not known
df = df.drop(columns=['TermReason', 'DateofTermination', 'EmpStatusID', 'EmploymentStatus'])
df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
# let's look at ManagerID next
df[df['ManagerID'].isnull()]
```

```python
# let's search for Webster Butler if maybe other columns have an ID
df[df['ManagerName']=='Webster Butler']
```

```python
# set ManagerID=39 where it's NaN
df.loc[df['ManagerID'].isnull(), 'ManagerID'] = 39.0
df[df['ManagerName']=='Webster Butler']
```

```python
# convert MarriedID and other boolean types
df['MarriedID'] = df['MarriedID'].astype(bool)
df['FromDiversityJobFairID'] = df['FromDiversityJobFairID'].astype(bool)
df['Termd'] = df['Termd'].astype(bool)
df['GenderID'] = df['GenderID'].astype(bool)
df.describe()
```

```python
# describe only works for numerical values
# we may inspect the types of the dataframe columns
df.dtypes
```

```python
# Employee_Name is not relevant for prediction and we have the ID for identification
df = df.drop(columns='Employee_Name')
df.dtypes
```

```python
# prediction library Scikit needs all predictors encoded numerically or booleans
# MaritalDesc is categorical and we have the same information in MaritalStatusID as int
# MarriedID holds a subset of data of MaritalStatusID
# let's drop MaritalDesc and MarriedID
df = df.drop(columns=['MarriedID', 'MaritalDesc'])
df.dtypes
```

```python
# GenderID isn't a helpful column name for a boolean
# column Sex tells us that GenderID=False is F (female) and True is M (male)
# let's drop Sex and rename GenderID
df = df.drop(columns='Sex')
df = df.rename(columns={'GenderID': 'isMale'})
df.dtypes
```

```python
# numerical DeptID holds the same information as Department
# let's drop Department
df = df.drop(columns='Department')
df.dtypes
```

```python
# numerical PerfScoreID holds the same information as PerformanceScore
# let's drop PerformanceScore
df = df.drop(columns='PerformanceScore')
df.dtypes
```

```python
# check Position's values to determine suitability
df['Position'].value_counts()
```

```python
# too many categories with too few entries, so we create a new boolean column
df['isProductTechnician'] = False
df.loc[df['Position']=='Production Technician I', 'isProductTechnician'] = True
df.loc[df['Position']=='Production Technician II', 'isProductTechnician'] = True
df
```

```python
# numeric PositionID holds the same information as Position
df = df.drop(columns=['Position', 'PositionID'])
df.dtypes
```

```python
df["State"].value_counts()
```

```python
# too many categories with too few entries
# create a new column isFromMA based on State and drop State
df['isFromMA'] = False
df.loc[df['State']=='MA', 'isFromMA'] = True
df = df.drop(columns='State')
df
```

```python
# let's drop ZIP as well as it would be only helpful within MA
df = df.drop(columns='Zip')
df
```

```python
# convert date strings to actual dates
df['DOB'] = pd.to_datetime(df['DOB'])
df['DOB'].value_counts()
```

```python
# some DOBs are in the future, due to a conversion error
# values 69–99 are mapped to 1969–1999, and values 0–68 are mapped to 2000–2068
from datetime import timedelta, date
df.loc[df['DOB'] > pd.Timestamp.now(), 'DOB'] -= timedelta(days=365.2425*100)
df['DOB'].value_counts()
```

```python
# numerical ManagerID holds the same information as ManagerName
# let's drop ManagerName
df = df.drop(columns='ManagerName')
df.dtypes
```

```python
# check distribution and correlation of CitizenDesc
df.groupby(['CitizenDesc', 'Termd']).size().unstack(fill_value=0)
```

```python
# only few employees are Non-Citizens but they terminate way more often
# we keep the variable as categorical
df['CitizenDesc'] = df['CitizenDesc'].astype('category')
df.dtypes
```

```python
# we remove the "race"-related columns HispanicLatino and RaceDesc
df = df.drop(columns=['HispanicLatino', 'RaceDesc'])
df.dtypes
```

```python
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df
```

```python
df.groupby(['RecruitmentSource', 'Termd']).size().unstack(fill_value=0)
```

```python
# many different observations but we'll use them as categories
# we can drop FromDiversityJobFairID as the information is included in here
df = df.drop(columns='FromDiversityJobFairID')
df['RecruitmentSource'] = df['RecruitmentSource'].astype('category')
df
```

```python
# LastPerformanceReview_Date has NaN values (no review)
# check how many NaNs and correlation to Termd
df.groupby(['LastPerformanceReview_Date', 'Termd']).size().unstack(fill_value=0)
```

```python
# all non-terminated employees had an interview (in 2019), all terminated employees didn't have one
# two possible interpretations:
# 1) Do performance review to keep employees from terminating!
# 2) All reviews happened 2019. Let's assume the company started reviews 2019 after so many people terminated.
# Following 2), we drop the column
df = df.drop(columns='LastPerformanceReview_Date')
df
```

## Build a Prediction Model

```python
# we use Scikit-Learn for modelling
# Scikit requires all variables to be numerical, so we need to encode the categorical columns
from sklearn.preprocessing import LabelEncoder
df.dtypes
```

```python
categoricals = ['CitizenDesc', 'RecruitmentSource']
encoder = LabelEncoder()
for i in categoricals:
    df[i] = encoder.fit_transform(df[i])
df.dtypes
```

```python
# What's the easiest prediction model you may think of?
# Just use the modus of Termd and you get a baseline accuracy.
len(df[df['Termd']==False]) / len(df)
```

```python
# import models from scikit learn
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
```

```python
# select the variables that we want to use to predict Termd
allPredictors = ['MaritalStatusID', 'DeptID', 'PerfScoreID', 'zSalary', 'ManagerID', 'SpecialProjectsCount', 'DaysLateLast30', 'Absences', 'isProductTechnician', 'isFromMA', 'age', 'tenure']
predictors = allPredictors

# fit a LogisticRegresssion as a first try
# this is called regression although we do classification (sorting entries into Termd: True/False)
model = LogisticRegression(solver='lbfgs', max_iter=1000).fit(df[predictors], df['Termd'])
# make prediction on training set
prediction = model.predict(df[predictors])

accuracy = metrics.accuracy_score(prediction, df['Termd'])
print("Accuracy: %s" % "{0:.3%}".format(accuracy))
```

```python
# select the variables that we want to use to predict Termd
predictors = ['MaritalStatusID', 'tenure', 'zSalary', 'RecruitmentSource']

# fit a decision tree
model = DecisionTreeClassifier().fit(df[predictors], df['Termd'])
# make prediction on training set
prediction = model.predict(df[predictors])

accuracy = metrics.accuracy_score(prediction, df['Termd'])
print("Accuracy: %s" % "{0:.3%}".format(accuracy))
```

```python
# sounds too good to be true, doesn't it?
# for now, we have used all data for training, testing and validation
# let's split by ~50/25/25%
train = df[0:150]
test = df[151:225]
validate = df[226:]
```

```python
# select the variables that we want to use to predict Termd
predictors = ['MaritalStatusID', 'tenure', 'zSalary', 'RecruitmentSource']

# fit a decision tree again - with training data
model = DecisionTreeClassifier().fit(train[predictors], train['Termd'])
# make prediction on training set
prediction = model.predict(train[predictors])

accuracy = metrics.accuracy_score(prediction, train['Termd'])
print("Accuracy on training data: %s" % "{0:.3%}".format(accuracy))
```

```python
# make prediction on test set
prediction = model.predict(test[predictors])

accuracy = metrics.accuracy_score(prediction, test['Termd'])
print("Accuracy on test data: %s" % "{0:.3%}".format(accuracy))
```

```python
# the accuracy on training data is 100% - the model fits perfectly to the data
# the accuracy on test data is worse than our baseline model - the trained model is overfitted on the training data
# could be changed by trying other (fewer) predictors or changing the model parameters

# fit a decision tree again - with training data and adjusted model parameters
model = DecisionTreeClassifier(max_depth=6, max_features='auto').fit(train[predictors], train['Termd'])
# make prediction on training set
prediction = model.predict(train[predictors])

accuracy = metrics.accuracy_score(prediction, train['Termd'])
print("Accuracy on training data: %s" % "{0:.3%}".format(accuracy))
```

```python
# make prediction on test set
prediction = model.predict(test[predictors])

accuracy = metrics.accuracy_score(prediction, test['Termd'])
print("Accuracy on test data: %s" % "{0:.3%}".format(accuracy))
```

```python
# more realistic but let's try a more sophisticated algorithm
# random forests are randomly generated decision trees that are combined into one
# different runs yield different results (due to randomness)

# use all variables for random forest to predict Termd
predictors = allPredictors

# fit a random forest (n randomized decision trees)
model = RandomForestClassifier(n_estimators=100).fit(train[predictors], train['Termd'])
# make prediction on training set
prediction = model.predict(train[predictors])

accuracy = metrics.accuracy_score(prediction, train['Termd'])
print("Accuracy on training data: %s" % "{0:.3%}".format(accuracy))
```

```python
# make prediction on test set
prediction = model.predict(test[predictors])

accuracy = metrics.accuracy_score(prediction, test['Termd'])
print("Accuracy on test data: %s" % "{0:.3%}".format(accuracy))
```

```python
# again the model is overfitted, let's find out which predictors to exclude
# random forests determine the feature importances
importance = pd.Series(model.feature_importances_, index=predictors).sort_values(ascending=False)
print(importance)
```

```python
# use most important variables for random forest to predict Termd
importantPredictors = ['tenure', 'zSalary', 'age', 'Absences', 'ManagerID']
predictors = importantPredictors

# fit a random forest (n randomized decision trees)
model = RandomForestClassifier(n_estimators=100).fit(train[predictors], train['Termd'])
# make prediction on training set
prediction = model.predict(train[predictors])

accuracy = metrics.accuracy_score(prediction, train['Termd'])
print("Accuracy on training data: %s" % "{0:.3%}".format(accuracy))
```

```python
# make prediction on test set
prediction = model.predict(test[predictors])

accuracy = metrics.accuracy_score(prediction, test['Termd'])
print("Accuracy on test data: %s" % "{0:.3%}".format(accuracy))
```

```python
# random forests can also be used to just get the feature importance and then use the predictors with other algorithms

# fit a random forest (n randomized decision trees) with adjusted model parameters
model = RandomForestClassifier(n_estimators=100, max_depth=5, warm_start=True).fit(train[predictors], train['Termd'])
# make prediction on training set
prediction = model.predict(train[predictors])

accuracy = metrics.accuracy_score(prediction, train['Termd'])
print("Accuracy on training data: %s" % "{0:.3%}".format(accuracy))
```

```python
# make prediction on test set
prediction = model.predict(test[predictors])

accuracy = metrics.accuracy_score(prediction, test['Termd'])
print("Accuracy on test data: %s" % "{0:.3%}".format(accuracy))
```

```python
# let's evaluate the model with our validation data set
prediction = model.predict(validate[predictors])
accuracy = metrics.accuracy_score(prediction, validate['Termd'])
print("Accuracy on validation data: %s" % "{0:.3%}".format(accuracy))
```

## Closing Remarks
* Using more complex models doesn't necessary yield better results.
* Feature engineering, feature selection and model parameter tuning can take long time.
* Sometimes, the data is just insufficient or does not correlate with the target variable at all, so no decent model may be trained.
* Different runs with models using randomness may yield different results.
* There are way more models than the three we discussed, try others.
